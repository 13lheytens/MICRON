```python
import requests
import re
import dill

def name2smiles(NDC):
    url3 = 'https://ndclist.com/?s=' + NDC
    r3 = requests.get(url3)
    name = re.findall('<td data-title="Proprietary Name">(.+?)</td>', r3.text)[0]
    
    url = 'https://dev.drugbankplus.com/guides/tutorials/api_request?request_path=us/product_concepts?q=' + name
    r = requests.get(url)
    drugbankID = re.findall('(DB\d+)', r.text)[0]

    url2 = 'https://www.drugbank.ca/drugs/' + drugbankID
    r2 = requests.get(url2)
    SMILES = re.findall('<dt class="col-md-2 col-sm-4">SMILES</dt><dd class="col-md-10 col-sm-8"><div class="wrap">(.+?)</div>', r2.text)[0]
    return SMILES

# idx2ndcPath = '../data/idx2ndc.pkl'
# idx2ndc = dill.load(open(idx2ndcPath, 'rb'))

# idx2drug = {}
for k, ndc in atc2ndc.values:
    if k not in list(data.keys()):
        for code in ndc:
            try:
                print (k, code)
                SMILES = name2smiles(code)
                print (k, SMILES)
                if k not in data:
                    data[k] = set()
                data[k].add(SMILES)
                if len(data[k]) >= 3:
                    break
            except:
                pass
# dill.dump(obj=idx2drug, file=open('idx2drug.pkl', 'wb'))
```
